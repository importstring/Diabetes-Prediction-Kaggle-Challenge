Gradient boosting is best for this problem.

Of the GBM methods, let's choose CatBoost bc:
- Categorical handling
- Great defaults
- Fast interference
It does have slower training, a smaller community, and few GPU support these aren't of concern.

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score

X = df.drop(['diagnosed_diabetes', 'id'], axis=1)
y = df['diagnosed_diabetes']
cat_features = ['gender', 'ethnicity', 'education_level',
                'income_level', 'smoking_status', 'employment_status']

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

oof_preds = np.zeros(len(X))
cv_scores = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = CatBoostClassifier(
        loss_function="Logloss",
        eval_metric="AUC",
        iterations=1500,
        learning_rate=0.05,
        depth=8,
        random_state=42,
        verbose=True
    )

    model.fit(
        X_train, y_train,
        cat_features=cat_features,
        eval_set=(X_val, y_val),
        early_stopping_rounds=100,
        verbose=False
    )

    val_pred = model.predict_proba(X_val)[:, 1]
    oof_preds[val_idx] = val_pred

    auc = roc_auc_score(y_val, val_pred)
    cv_scores.append(auc)
    print(f"Fold {fold} AUC: {auc:.5f}")

print(f"Mean CV AUC: {np.mean(cv_scores):.5f}")

# train final model on ALL data (no split)
model = CatBoostClassifier(
    loss_function="Logloss",
    eval_metric="AUC",
    iterations=1500,
    learning_rate=0.05,
    depth=8,
    random_state=42,
    verbose=True,
)
model.fit(X, y, cat_features=cat_features, verbose=False)
```

Taking a look at what the competetion used and their results. Yup everyone used Catboost and yes 68 is good

```python
submission_df = pd.read_csv('data/test.csv')
X_submit = submission_df[X.columns]
probs = model.predict_proba(X_submit)

diabetes_prob = probs[:, 1]
```

Return the output in the requested format

```python
submission = pd.DataFrame({
    "patient_id": submission_df["id"],
    "diagnosed_diabetes": diabetes_prob
})
submission.to_csv("submission.csv", index=False)
```
